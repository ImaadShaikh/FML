```python
#Program to compute pay using a function
def computePay(hours):
    if(hours>40):
        k=hours-40
        pay=k*15+40*10
    else:
        pay=hours*10
    return pay

hours=int(input("Enter hours:"))
computePay(hours)
```

```python
#Program to find smallest number
def smallNum(a):
    small=9999
    for i in a:
        if i<small:
            small=i
    return small

smallNum([9,42,12,3,74,15])
```

```python
def sum(x,y,z=None):
    if(z==None):
        return x+y
    else:
        return x+y+z

sum(2,3,4)
```
